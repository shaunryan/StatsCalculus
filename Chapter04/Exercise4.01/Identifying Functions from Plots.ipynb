```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)

plt.plot(
    x,  # evenly spaced numbers in the x-axis
    np.ones(1000) * 2  # all 2s in the y-axis
)

plt.show()
```

```python
plt.plot(x, x)
plt.show()
```

```python
plt.plot(x, x ** 2)
plt.show()
```
